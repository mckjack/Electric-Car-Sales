```python
import psycopg2
import pandas as pds
from sqlalchemy import create_engine
```

```python
# PGHOST = 'postgres_endpoint'
PGHOST = 'final-project-data-analytics.c53bsxseplpp.us-east-2.rds.amazonaws.com'

# PGDATABASE = 'postgres_db_name'
PGDATABASE = 'AWS_RDS_Final_Project_db'

# PGUSER = 'postgres_user'
PGUSER = 'root'

# PGPASSWORD = 'postgres_db_password'
PGPASSWORD = 'abcd1234'
```

```python
# Create an engine instance

# alchemyEngine   = create_engine('postgresql+psycopg2://root:abcd1234@final-project-data-analytics.c53bsxseplpp.us-east-2.rds.amazonaws.com:5432/AWS_RDS_Final_Project_db', pool_recycle=3600);
conn_string =  "postgresql+psycopg2://"+ PGUSER + ":" + PGPASSWORD + "@"+ PGHOST + ":5432/"+ PGDATABASE

alchemyEngine = create_engine(conn_string, pool_recycle=3600)

# Connect to PostgreSQL server
dbConnection = alchemyEngine.connect()

print("AWS RDS Postgres DB Connected!")
```

```python
# Read data from PostgreSQL database table and load into a DataFrame instance

brent_oil_prices_df = pds.read_sql("select * from \"brent_oil_prices\"", dbConnection)

pds.set_option('display.expand_frame_repr', False)

# Print the DataFrame

print("brent_oil_prices_df")
brent_oil_prices_df.head()

 
```

```python
# Check data types of DF
brent_oil_prices_df.dtypes
```

```python
# Change the data type of "date" column from object to datetime
brent_oil_prices_df['date'] = pds.to_datetime(brent_oil_prices_df['date'])
brent_oil_prices_df.head()
```

```python
brent_oil_prices_df.dtypes
```

```python
# calculate the average price by month
brent_oil_prices_df = brent_oil_prices_df.resample('M', on="date").mean()
brent_oil_prices_df.head()
```

```python
brent_oil_prices_df = brent_oil_prices_df.reset_index()
brent_oil_prices_df.head()
```

```python
# Only show the year and month
brent_oil_prices_df['date']= brent_oil_prices_df['date'].dt.to_period('M')
brent_oil_prices_df.head()
```
